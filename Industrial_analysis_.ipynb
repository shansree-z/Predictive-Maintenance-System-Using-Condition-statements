```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("stephanmatzka/predictive-maintenance-dataset-ai4i-2020")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

# 1. Identify the correct file path
# path was defined in your first cell by kagglehub
files = os.listdir(path)
csv_file = [f for f in files if f.endswith('.csv')][0] # Finds the first CSV file
csv_path = os.path.join(path, csv_file)

print(f"Loading dataset: {csv_file}")
df = pd.read_csv(csv_path)

# 2. The If-Else Industry Logic Engine
def predict_failure(row):
    # Using the exact column names from the AI4I 2020 dataset
    temp_diff = row['Process temperature [K]'] - row['Air temperature [K]']
    wear = row['Tool wear [min]']
    torque = row['Torque [Nm]']

    if wear > 240:
        return "CRITICAL: Tool Wear Limit"
    elif temp_diff < 8.6 and row['Rotational speed [rpm]'] < 1380:
        return "CHECK: Heat Dissipation"
    elif torque > 65:
        return "WARNING: Overstrain Detected"
    else:
        return "OPERATIONAL: System Healthy"

# 3. Apply the logic to create a maintenance report
df['Maintenance_Action'] = df.apply(predict_failure, axis=1)

# Display the "Big Solution" result
print("\n--- FACTORY FLOOR MAINTENANCE REPORT ---")
print(df['Maintenance_Action'].value_counts())
print("---------------------------------------")
```

```python
# Real-time data from one machine sensor
machine_id = "CNC-001"
tool_wear = 245        # Minutes of use
temperature_gap = 7.2  # Difference between machine and air temp
torque_load = 45       # Physical strain level

# The Small Solution: Simple If-Else Logic
if tool_wear > 240:
    status = "STOP: Tool Wear Limit Reached"
    action = "Replace cutting head immediately."

elif temperature_gap < 8.6:
    status = "WARNING: Overheating"
    action = "Check cooling system and filters."

elif torque_load > 65:
    status = "ALERT: High Strain"
    action = "Reduce motor speed."

else:
    status = "HEALTHY"
    action = "Continue production."

# Output the Result
print(f"Machine: {machine_id}")
print(f"Status: {status}")
print(f"Action: {action}")
```
